**Содержание**

Данный датасет включает в себя:

Информацию о 49 лигах с сезона 1999 года по 2020 год. Данные о 11к игроках, такие как: дата рождения, рост, вес, национальность, место учебы. А так же статистику за сезон: бросковые данные (двух-, трехочковые, штрафные), подборы, блоки, ассисты(пасы), минуты на площадке и тд.

- League = League Name / Наименование лиги
- Season = Season Year YYYY-YYYY / Сезон такого то года в формате YYYY-YYYY
- Stage = International, NBA:Playoffs,Regular_Seasion / Формат турнира - международный, нба: регулярные сезон или плейофф
- Player = Player Full Name / полное имя игрока
- Team = Team Name / название команды
- GP = Games Played / игр сыграно
- MIN = Minutes Played / проведенные минуты на площадке
- FGM = Field Goals Made / реализованные броски
- FGA = Field Goals Attempts / попыток бросков
- 3PM = Three Points Made / реализованные 3хочковые
- 3PA = Three Points Attempts / попытки 3хочоквых
- FTM = Free Throws Made / реализованные штрафные
- FTA = Free Throws Attempts / попытки штрафных
- TOV = Turnovers / потери
- PF = Personal Fouls / персональные фолы
- ORB = Offensive Rebounds / подборы в нападении
- DRB = Defensive Rebounds / подборы в защите
- REB = Rebounds / общее число подборов
- AST = Assists / передачи
- STL = Steals / кража мяча
- BLK = Blocks / заблокированные мячи
- PTS = Points / количество очков
- birth_year = Birth Year / год рождения
- birth_month = Birth Month / месяц рождения
- birth_date = Birth Date / дата рождения
- height = Height (Feet) / рост в футах
- height_cm = Height (CM) / рост в сантиметрах
- weight = Weight (Pounds) / вес в пудах
- weight_kg = Weight (KG) / вес в кг
- nationality = Nationality / национальность
- high_school = High School / место обучения
- draft_round / в каком раунде выбрали игрока на драфте
- draft_pick / под каким номером игрока выбрали
- draft_team / какая команда задрафтовала игрока

b) Points vs. Attempts

```python
def add_cols(df, *args):
    cols_range = range(1, len(list(args)))
    col_list = [col for col in args]
    base_col = df[col_list[0]].copy()

    for i in cols_range:
        base_col += df[col_list[i]]
    return base_col
```

```python
point_leaders['Total_Attempts'] = add_cols(point_leaders, 'Field_Goals_Attempts', 
                                           'Three_Points_Attempts', 'Free_Throws_Attempts')
```

```python
point_leaders['Total_Throws_Made'] = add_cols(point_leaders, 'Field_Goals_Made', 
                                              'Three_Points_Made', 'Free_Throws_Made')
```

```python
point_leaders['Points_Attempt_Ratio'] = (point_leaders['Points']/point_leaders['Total_Attempts'])
```

```python
size=point_leaders['Games_Played']
test_5 = px.scatter(
    point_leaders, x='Points', y='Points_Attempt_Ratio',
    color='Points_Attempt_Ratio',
    color_discrete_sequence=px.colors.sequential.Rainbow,
    size=size, 
    hover_name='Player',
    title='Количество очков в зависимости от процента реализации',
    trendline='ols',
    template=temp
)
test_5.update_traces(marker=dict(
    sizeref=2*max(size)/(40**2), 
    sizemode='area', 
    sizemin=0,
))
test_5.update_layout(
    coloraxis_colorbar_thicknessmode='pixels',
    coloraxis_colorbar_thickness=10,
#     coloraxis_showscale=False
)
test_5.update_traces(
    line_dash='dot',
    line_width=4,
    line_color='royalblue')

test_5.show()
```

c) Skill Profile (Line-Polar Viz)

```python
from sklearn.preprocessing import MinMaxScaler

min_max_scaler= MinMaxScaler()
```

```python
skills = ['Rebounds', 'Assists', 'Steals', 'Blocks', 'Points']
```

```python
player_profile = nba.groupby('Player').sum().sort_values(by='Points', ascending=False).reset_index()
```

```python
normalized_skills = min_max_scaler.fit_transform(player_profile[skills])
```

```python
normalized_skills_df = pd.DataFrame(normalized_skills, columns=player_profile[skills].columns)
normalized_skills_df = normalized_skills_df.add_suffix('_Normalized')
```

```python
player_skills = pd.concat([player_profile.loc[:,['Player', 'Points']], normalized_skills_df], axis=1)

player_skills = player_skills.sort_values(by='Points', ascending=False).head(20)
player_skills = player_skills.drop('Points', axis=1)
```

```python
polar_df = pd.melt(player_skills, id_vars=['Player'], value_vars=player_skills.drop('Player', axis=1).columns, var_name='Skills')
```

```python
test_6 = px.line_polar(polar_df, r="value", theta="Skills", color='Player', line_close=True,
                    color_discrete_sequence=px.colors.sequential.Plasma_r,
                    template="plotly_dark", 
                    animation_frame="Player",
                    animation_group="Player",
                    title='Выбрать игрока для отображения пятиугольника навыков:'
                   )
test_6.show()
```

d) Skill Profile (Bar-Polar Viz)

```python
test_7 = px.bar_polar(polar_df, r="value", theta="Skills", color='value',
                    color_continuous_scale=px.colors.sequential.Viridis,
                    animation_frame="Player",
                    animation_group="Player",
                    template="plotly_dark", 
                    title='')
test_7.update_layout(
    coloraxis_colorbar_thicknessmode='pixels',
    coloraxis_colorbar_thickness=10,
    coloraxis_colorbar_len=1.5
)
test_7.show()
```
